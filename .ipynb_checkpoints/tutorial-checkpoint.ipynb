# Como rodar uma query SQL no python e obter a resposta como um DataFrame

```python
# conexao com o client
from google.cloud.bigquery import Client, QueryJobConfig
client = Client()
```

```python
query = """SELECT * FROM `hacka-dados.hacka_dados.tabela_dados_hacka` """
job = client.query(query)
df = job.to_dataframe()
```

```python
df.head()
```

# Escrevendo um df no banco de dados

1) eh necessario um criar um schema ou utilizar um ja criado -- CREATE SCHEMA NOME;

```python
SCHEMA_name = 'NomeSchema'
query = """ CREATE SCHEMA {0}""".format(SCHEMA_name)
job = client.query(query)
```

```python
table_id = SCHEMA_name + '.tabela_analise1'
dataframe = df
```

```python
query = "DROP TABLE IF EXISTS {0}".format(table_id)
job = client.query(query)
```

```python
job = client.load_table_from_dataframe(
    dataframe, table_id
)  # Make an API request.
job.result()  # Wait for the job to complete.
```

```python
table_id
```

```python
query = """SELECT * FROM `NomeSchema.tabela_analise1` LIMIT 1000"""
job = client.query(query)
df = job.to_dataframe()
```

```python
df.head()
```

Voce pode verificar os dados na aba `BigQuery In Notebooks` a esquerda

# Treinando um modelo simples

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
query = """SELECT * FROM `hacka-dados.hacka_dados.tabela_dados_hacka` """
job = client.query(query)
df = job.to_dataframe()
```

```python
X = df[['Capacidade_de_Pagamento_7','tempo_desde_primeiro_contrato','tempo_desde_ultimo_contrato','QTD_OUTROS','VrFinanciamento']].fillna(0)
y = df.fl_renovou.values
clf = LogisticRegression(random_state=0).fit(X, y)
y_hat1 = clf.predict(X)
y_hat2 =  clf.predict_proba(X)
print(classification_report(y, y_hat1 ))
```

```python
X = df[['Capacidade_de_Pagamento_7','tempo_desde_primeiro_contrato','tempo_desde_ultimo_contrato','QTD_OUTROS','VrFinanciamento']].fillna(0)
y = df.fl_renovou.values
clf = LogisticRegression(random_state=0,class_weight='balanced').fit(X, y)
y_hat1 = clf.predict(X)
y_hat2 =  clf.predict_proba(X)
print(classification_report(y, y_hat1 ))
```

# Visualizando os resultados dados

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X = df[['Capacidade_de_Pagamento_7','tempo_desde_primeiro_contrato','tempo_desde_ultimo_contrato','QTD_OUTROS','VrFinanciamento']].fillna(0)
```

# PCA

```python
pca = PCA(n_components=2)
Xp = pca.fit_transform(X)
sns.scatterplot( x=Xp[:,0], y=Xp[:,1], hue=y_hat2[:,1] )
```

```python
sns.scatterplot( x=Xp[:,0], y=Xp[:,1], hue= y )
```

# Salvando figuras na sua maquina local

```python
# esse codigo gera uma figura no ambiente GCP
img = plt.figure(figsize=(8, 6))
sns.scatterplot( x=Xp[:,0], y=Xp[:,1], hue=y )
plt.savefig("fig1.png")
```

1) clicar na figura gerada na aba `File Browser` a esquerda 
2) abrir a imagem (clicar com o botao esquerdo) 
3) clicar com o botao direito e clicar em salvar imagem como (selecionar o local no PC)

# TSNE

```python
X_embedded = TSNE(n_components=2, learning_rate='auto', init='random').fit_transform(X.iloc[:10000])
```

```python
img = plt.figure(figsize=(8, 6))
sns.scatterplot( x=X_embedded[:,0], y=X_embedded[:,1], hue=y_hat2[:10000,1] )
plt.show()
img = plt.figure(figsize=(8, 6))
sns.scatterplot( x=X_embedded[:,0], y=X_embedded[:,1], hue=y[:10000] )
plt.show()
```
